# Quantum States and Gates

## Qubit States

       
  A Quantum Circuit comprise of **Qubits** and **Quantum Gates**.
  
Qubit or quantum bit is the basic unit of quantum information.
  
A single qubit quantum state can be written as  
$|ψ⟩=α|0⟩+β|1⟩$, where $α$ and $β$ are complex numbers.    
  
During measurement, the probability of the bit being in $|0⟩$ is $|α|^2$ and $|1⟩$ is $|β|^2$. 
By conservation of probability- $|α|^2+|β|^2=1 $.  
Therefore we can also represent it in vector form: $|ψ⟩ = \begin{pmatrix} α\\β\end{pmatrix} $
    


```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import array_to_latex, plot_histogram,plot_bloch_vector, plot_bloch_multivector, plot_state_qsphere
from math import sqrt, pi
```

```python
# simulator 
sim = Aer.get_backend('aer_simulator')  
```

```python
# Create a quantum circuit with one qubit
qc = QuantumCircuit(1) 
```

$|q⟩ = \alpha|0⟩ + \beta|1⟩$  
$\alpha, \beta \in C$

#### Default State
$|q⟩ = |0⟩ =  \begin{pmatrix} 1\\0\end{pmatrix} $

#### State initialization

$|q⟩ = \frac{1}{\sqrt2}|0⟩ + \frac{i}{\sqrt2}|1⟩$

```python
initial_state = [1/sqrt(2), 1j/sqrt(2)]   # Define initial_state as |1>
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.draw('mpl')  # view circuit
```

```python
qc.save_statevector()
qobj = assemble(qc)     # Create a Qobj from the circuit for the simulator to run
result = sim.run(qobj).result() # Do the simulation and return the result

out_state = result.get_statevector()
array_to_latex(out_state, prefix="\\text{Statevector} = ") # Display the output state vector
```

```python
counts = result.get_counts()  ## measuring output before collapsing
print(counts)
```

## Single Qubit Gates

#### Quantum Operations
Quantum gates/operations are usually represented by matrices.   
A gate which acts on 1 qubit is represented by a 2×2 unitary matrix $U$.    
$|ψ′⟩=U|ψ⟩$

###### The X-Gate  
$X = \begin{bmatrix}0 &1\\1&0\end{bmatrix} = |0⟩⟨1| + |1⟩⟨0| $

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))

plot_bloch_multivector(state)
```

#### Bloch sphere representation of a qubit state:    
$|q⟩=cos\frac{θ}{2}|0⟩+e^{iϕ}sin\frac{θ}{2}|1⟩$    
θ, ϕ are real numbers; 0≤ϕ<2π, and 0≤θ≤2π.    
  
From this, it is clear that there is a one-to-one correspondence between qubit states ($C^2$) and the points on the surface of a unit sphere ($R^3$). 

###### The Y-Gate  
$Y = \begin{bmatrix}0 &-i\\i&0\end{bmatrix} = -i|0⟩⟨1| + i|1⟩⟨0| $

```python
qc = QuantumCircuit(1)

qc.y(0)
qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))

plot_bloch_multivector(state)
```

###### The Z-Gate  
$Z = \begin{bmatrix}1 &0\\0&-1\end{bmatrix} = |0⟩⟨0| + -|1⟩⟨1| $

```python
qc = QuantumCircuit(1)
qc.x(0)
qc.z(0)
qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))

plot_bloch_multivector(state)
```

###### The U Gate 
$U(θ,\phi,λ) = \begin{bmatrix}cos(θ/2) &-e^{iλ}sin(θ/2)\\e^{i\phi}sin(θ/2)&e^{i(\phi+λ)}cos(θ/2)\end{bmatrix}$; where θ,$\phi$,λ are real numbers.  
Any gate can be obtained using U gate.  
  
$U(\pi/2,0,\pi) = \frac{1}{\sqrt2}\begin{bmatrix}1 &1\\1&-1\end{bmatrix} = H $  
$U(0,\phi,0) = \begin{bmatrix}1 &0\\0&e^{i\phi}\end{bmatrix} =P(λ) $,

```python
qc = QuantumCircuit(1)
#qc.x(0)
qc.u(pi/4, pi/2, -pi/2, 0)
#qc.u(pi/2, -pi/6, 0, 0)
qc.draw('mpl')

```

```python
qc.save_statevector()
qobj = assemble(qc)
state = sim.run(qobj).result().get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))

plot_bloch_multivector(state)
```

## Multi-Qubit States

$|a⟩ = \begin{bmatrix}a_0\\a_1\end{bmatrix}$, $|b⟩ = \begin{bmatrix}b_0\\b_1\end{bmatrix}$  

Then State of $|ba⟩ = |b⟩⊗|a⟩=\begin{bmatrix}b_0a_0\\b_0a_1\\b_1a_0\\b_1a_1\end{bmatrix}$

```python
qc = QuantumCircuit(3)
# Apply H-gate to each qubit:
for qubit in range(3):
    qc.h(qubit)

qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
result = sim.run(qobj).result()
state = result.get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))
plot_bloch_multivector(state)
```

```python
plot_state_qsphere(state)
```

```python
counts = result.get_counts()
print(counts)
plot_histogram(counts)
```

## Multi-Qubit Gates

###### The CNOT-gate  
This gate is a conditional gate that performs an X-gate on a target second qubit, only if the state of the control qubit is |1⟩.

```python
qc = QuantumCircuit(2)
#qc.x(0)
qc.cx(0,1)# CNOT

qc.draw('mpl')
```

Output of q1 becomes q0 xor q1.

```python
Output of q1 becomes q0 xor q1.qc.save_statevector()
qobj = assemble(qc)
result = sim.run(qobj).result()
state = result.get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))
```

```python
qc.save_unitary()
qobj = assemble(qc)
unitary = sim.run(qobj).result().get_unitary()

# Display the operation using matrix:
array_to_latex(unitary, prefix="\\text{Circuit = } ")
```

## Implications of CNOT gate

### Entangled State

$CNOT |01⟩ = \frac{1}{\sqrt2}(|00⟩ + |11⟩)$  
This is also called the bell-state. When this state is measured we have,  
50% probability of being measured in the state |00⟩, and 50% chance of being measured in the state |11⟩.   
As a consequence, 0% chance being in |01⟩ and |10⟩.

```python
qc = QuantumCircuit(2)
# Apply H-gate to the first:
qc.h(0)

# Apply a CNOT:
qc.cx(0,1)

qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
result = sim.run(qobj).result()
state = result.get_statevector()
display(array_to_latex(state, prefix="\\text{Out State} = "))
```

```python
plot_histogram(result.get_counts())
```

```python
plot_bloch_multivector(state)
```

We cannot visualize the correlation between the 2 qubits using bloch-sphere.

```python
plot_state_qsphere(state)
```

$CNOT |++⟩ = \frac{1}{2}(|00⟩ + |01⟩ + |10⟩ + |11⟩)$

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.draw('mpl')
```

```python
qc.save_statevector()
qobj = assemble(qc)
final_state = sim.run(qobj).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Out Statevector} = "))
plot_bloch_multivector(final_state)
```

### Phase Kickback  
Not operation effects back the controller qubit, instead of effecting the target.  
The phase of the control quit changes, when it is in superposition state.

$CNOT |-+⟩ =  |--⟩$

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.h(1)
qc.cx(0,1)
display(qc.draw('mpl'))
```

```python
qc.save_statevector()
qobj = assemble(qc)
final_state = sim.run(qobj).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Out Statevector} = "))
plot_bloch_multivector(final_state)
```
